# API Used: NYTimes Top Stories (Current Top 38 Articles in the Arts Section)

```python
import numpy as np
import pandas as pd
import requests
import json

# Requests the API data and reads the data into a Python dict
site = "https://api.nytimes.com/svc/topstories/v2/arts.json?api-key=xFgGIzF7OySkp9Z3GIW4iO22HSdnwbnvxF9UfAogc0YA1kbM "
nyt = requests.get(site)
list_data = json.loads(nyt.text)

print(list_data)
```

### Data is messy. It needs to be filtered to find the needed info before converting.

```python
# To find which key contains article data
print(list_data.keys())
print("\n")

# Finds how many articles are in the list
print(list_data['num_results'])
print("\n")

# Returns a list of all article info
results = list_data['results']
print(results)
```

### Results is a list of dicts, so it can be sent directly into the DataFrame without further cleaning
### Designated keys from the results are used for the index and columns

```python
df = pd.DataFrame(results, columns=['section', 'subsection', 'title', 'byline', 'item_type', 'updated_date','published_date'])
df.set_index('updated_date')
```
