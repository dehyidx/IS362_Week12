# API Used: NYTimes Books (Current Top 15 Hardcover Fiction List)

```python
import numpy as np
import pandas as pd
import requests
import json

# Requests the API data and reads the data into a Python dict
site = "https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key=JBLukRWgCCmxmPSNzGPQGrmBwQ7Gqt4Z"
nyt = requests.get(site)
list_data = json.loads(nyt.text)

print(list_data)
```

### Data is messy. It needs to be filtered to find the needed info before converting.

```python

print(list_data.keys())
print("\n")

# Finds how many books are in the list
print(list_data['num_results'])
print("\n")

# Returns a dict of all book info
print(list_data['results'])

```

### Results is a nested dict, so we can use .keys() to find the 'books' dict.

```python
print(list_data['results'].keys())
print('\n')
print(list_data['results']['books'])

```

### From there, a tidy DataFrame can be made using chosen 'books' keys for the index and columns ('books is a list of dicts).

```python
df = pd.DataFrame(list_data['results']['books'], columns=['rank', 'title', 'author', 'publisher'])
df.set_index('rank')
```
